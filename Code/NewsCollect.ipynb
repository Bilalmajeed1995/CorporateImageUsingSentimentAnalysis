```python
import requests
from GoogleNews import GoogleNews
import pandas as pd
import time
from newspaper.article import ArticleException, ArticleDownloadState
from newspaper import Article
import datetime
```

```python
search_lst1 = ["Apple", "Microsoft", "GameStop", "Mercedes-Benz", "Tesla",
               "Toyota", "Nike", "Adidas", "Walmart", "Target", "Home Depot"]
search_lst2 = ["Best Buy", "Costco", "Hyatt Hotels", "Marriott", "Verizon", "Rogers Communications", 
               "AT&T", "Citigroup", "JPMorgan Chase", "Wells Fargo Bank"]
search_lst3 = ["McDonalds", "Cineplex", "Starbucks", "Berkshire Hathaway", "Delta Airlines",
               "State Farm Insurance", "Allstate", "CVS Healthcare"]
```

```python
googlenews_df1 = pd.DataFrame()

for search in search_lst1:
    googlenews = GoogleNews(start='01/01/2020',end='12/31/2020')
    googlenews.search(search)
    
    result = googlenews.result()
    temp_df = pd.DataFrame(result)
    temp_df["search"] = search
    googlenews_df1 = googlenews_df1.append(temp_df)

    for i in range(2,4):
        googlenews.getpage(i)
        result = googlenews.result()
        temp_df = pd.DataFrame(result)
        temp_df["search"] = search
        googlenews_df1 = googlenews_df1.append(temp_df)
        time.sleep(10)
    
    for i in range(6,9):
        googlenews.getpage(i)
        result = googlenews.result()
        temp_df = pd.DataFrame(result)
        temp_df["search"] = search
        googlenews_df1 = googlenews_df1.append(temp_df)
        time.sleep(10)
    
    googlenews.getpage(10)
    result = googlenews.result()
    temp_df = pd.DataFrame(result)
    temp_df["search"] = search
    googlenews_df1 = googlenews_df1.append(temp_df)
    
    time.sleep(60)

googlenews_df1 = googlenews_df1.reset_index()
googlenews_df1 = googlenews_df1.drop(columns = ['index', 'img', 'datetime'])
```

```python
googlenews_df.to_csv('../Data/raw_data/googlenews_links.csv', index=False, header=True)
```

```python
links = googlenews_df[['link', 'search']]
links = links.values.tolist()
```

```python
df_google = pd.DataFrame()

for i in range(0, len(links)):
    try:
        article_google = Article(links[i][0])
        article_google.download()
        article_google.parse()
        article_google.nlp()
        article_info = {'company': [links[i][1]],
                        'link': [links[i][0]],
                        'title' : [article_google.title],
                        'nlp_summary': [article_google.summary],
                        'publish date': [datetime.datetime.date(article_google.publish_date)]}
        temp_df = pd.DataFrame(article_info)
        df_google = df_google.append(temp_df)
        time.sleep(2)
    
    except:
        article_info = {'company': [links[i][1]],
                        'link': [links[i][0]],
                        'title': ["null"],
                        'nlp_summary': ["null"],
                        'publish date': ["null"]}

        temp_df = pd.DataFrame(article_info)
        df_google = df_google.append(temp_df)
        continue
```

```python
df_google.to_csv('../Data/google_articles.csv', index=False, header=True)
```
