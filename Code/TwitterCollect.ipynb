```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import time
```

```python
tag_lst = ["#Apple", "#Microsoft", "#GameStop", "#MercedesBenz", "#Tesla",
           "#Toyota", "#Nike", "#Adidas", "#Walmart", "#Target", "#HomeDepot", 
           "#BestBuy", "#Costco", "#Hyatt", "#Marriott", "#Verizon", "#RogersCommunications", 
           "#AT&T", "#Citi", "#JPMorgan", "#WellsFargo", "#McDonalds", 
           "#Cineplex", "#Starbucks", "#BerkshireHathaway", "#DeltaAirlines",
           "#StateFarm"]
```

```python
# Creating list to append tweet data to
tweets_list = []

# Using TwitterSearchScraper to scrape data and append tweets to list (Q1)
for tag in tag_lst:
    search = tag + " since:2020-01-01 until:2020-03-31 lang:en -filter:retweets"
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(search).get_items()):
        if i > 1250:
            break
        tweets_list.append([tag, tweet.date, tweet.id, tweet.content])
    time.sleep(60)
        
tweets_df1 = pd.DataFrame(tweets_list, columns=['Tag', 'Datetime', 'Tweet Id', 'Text'])
```
